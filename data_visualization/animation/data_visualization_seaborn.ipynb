```python
import seaborn as sns
sns.__version__
```

```python
print(sns.get_dataset_names())
```

```python
# load dataset
df = sns.load_dataset("tips")
print(df.head())
print(df.shape)
print(df.std)
```

```python
import matplotlib.pyplot as plt
# basic graph
sns.lineplot(data=df, x='total_bill', y='tip')
plt.title("Seaborn line plot example")
plt.show()
```

```python
# relative plot
# split Lunch and Dinner time
sns.relplot(data=df, x='total_bill', y='tip', 
#             kind="scatter", 
            kind="line",
            col='time',    #  column by time
            hue="smoker",  
            style="smoker",
            size="size",   # tip size 
           ) 
#plt.title("Seaborn line plot example")
plt.show()
```

```python
# apply themme
sns.set_theme()
sns.relplot(data=df, x='total_bill', y='tip', 
            kind="scatter", 
            col='time',    #  column by time
            hue="smoker",  
            style="smoker",
            size="size",   # tip size 
           ) 
#plt.title("Seaborn line plot example")
plt.show()
```

```python
# linear regression 
sns.set_theme()
sns.lmplot(data=df, x='total_bill', 
            y='tip', 
            col='time',    #  column by time
            hue="smoker"
           ) 
plt.show()
```

```python
# displot 
sns.set_theme()
sns.displot(data=df, x='total_bill', 
            col='time',
            kde=True    # kernel distribution line
           ) 
plt.show()
```

```python
sns.set_theme()
sns.displot(data=df, 
            x='total_bill', 
            col='time',
            kind='ecdf',
            hue='smoker',
            rug=True,
            rug_kws={'color': "r", 
                     'alpha': 0.4, 
                     'linewidth': 2,
                     'height':0.2
                    }
           ) 
plt.show()
```

```python
# categorical plot
sns.set_theme()
sns.catplot(data=df, 
            x='day',    # by day category
            y='total_bill',
            kind='swarm',
            hue='smoker'
           ) 
plt.show()
```

```python
# violin plot
sns.set_theme()
sns.catplot(data=df, 
            x='day',    # by day category
            y='total_bill',
            kind='violin',
            hue='smoker'
           ) 
plt.show()
```

```python
# split violin plot
sns.set_theme()
sns.catplot(data=df, 
            x='day',    # by day category
            y='total_bill',
            kind='violin',
            hue='smoker',
            split=True
           ) 
plt.show()
```

```python
# bar plot
sns.set_theme()
sns.catplot(data=df, 
            x='day',    # by day category
            y='total_bill',
            kind='bar',
            hue='smoker'
           ) 
plt.show()
```

```python
# box plot
sns.set_theme()
sns.boxplot(data=df, 
            x=df['total_bill']
           ) 
plt.show()
```

```python
# count plot
sns.set_theme()
sns.countplot(data=df, 
            x='sex'
           ) 
plt.show()
```

```python
# strip plot for distributions
sns.set_theme()
sns.stripplot(data=df, 
            x='total_bill'
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, 
            x='total_bill',
            y='day'
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, 
            x='day',
            y='total_bill',
            hue='day'
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, 
            x='day',
            y='total_bill',
            hue='sex'
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, 
            x='day',
            y='total_bill',
            hue='size'
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, 
            x='day',
            y='total_bill',
            hue='size',
            palette='deep'
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, 
            x='day',
            y='total_bill',
            hue='size',
            dodge=True,
            jitter=False
           ) 
plt.show()
```

```python
# strip plot for distributions, additional value on y
sns.set_theme()
sns.stripplot(data=df, orient='h') 
plt.show()
```
