```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.linear_model import LinearRegression 
import random
```

```python
# create a regression line
x_values = []
y_values = []
```

```python
# create scatter plot
for _ in range(5):
    x_values.append(random.randint(0,100))
    y_values.append(random.randint(0,100))
    
    plt.xlim(0,100)
    plt.ylim(0,100)
    plt.scatter(x_values, y_values, color='black')
    plt.pause(0.01) 

plt.show()
```

```python
# plot regression line 
reg = LinearRegression()

for i in range(50):
    plt.clf()
    x_values.append(random.randint(0,100))
    y_values.append(random.randint(0,100))
    
    # make np array
    x = np.array(x_values)
    x = x.reshape(-1, 1)
    y = np.array(y_values)
    y = y.reshape(-1, 1)
    
    if i % 5 == 0:
        reg.fit(x, y)
        plt.xlim(0,100)
        plt.ylim(0,100)
        plt.scatter(x_values, y_values, color='black')
        plt.plot(list(range(100)), reg.predict(np.array([x for x in range(100)]).reshape(-1,1)))
        plt.pause(0.01) 

plt.show()

```

```python
# create bar plot 
import matplotlib.pyplot as plt 
import random 

values = [0] * 50

for i in range(10):
    plt.clf()
    values[i] = random.randint(0,100)
    plt.xlim(0,50)
    plt.ylim(0,100)
    plt.bar(list(range(50)), values)
    plt.pause(0.001)

plt.show()
    
```

```python
# interactive plot with ipywidgets
import ipywidgets 
import numpy as np 
import matplotlib.pyplot as plt 

X = np.random.uniform(0, 5, size=100)
noise = np.random.normal(size=100)

def plot_fct(w=1):
    y = 2 * X + w * noise 
    plt.scatter(X, y)
```

```python
ipywidgets.interact(plot_fct, w=(0, 5, 0.1))
```

```python
from sklearn.datasets import make_moons

def plot_moons(samples=200, noise=0):
    moons = make_moons(n_samples=samples, noise=noise, random_state = 50)
    X, y = moons[0], moons[1]

    plt.scatter(X[:,  0], X[:, 1],  c=y)
```

```python
ipywidgets.interact(plot_moons, samples=[200, 500, 1000], noise=(0, 2, 0.025))
```

```python
# plot_sin 
def plot_sin(start=0, end=30, factor=1, grid=False, plot_cos=False):
    X = np.linspace(start, end, (end-start) * 10)
    y = np.sin(X) * factor
    plt.grid(grid)
    plt.plot(X, y)
    if plot_cos:
        y = np.cos(X)
        plt.plot(X, y)
```

```python
ipywidgets.interact(plot_sin, start=(0, 10, 1), end=(20, 50, 1), factor=(0,5,0.1), grid=False)
```

```python
import math 

def plot_sigmoid(x_in = 0):
    X = np.linspace(-5, 5, 1000)
    y = 1 / (1 + np.exp(-X))
    y_in = 1 / (1 + math.exp(-x_in))
    
    plt.plot(X, y)
    plt.scatter(x_in, y_in, c="r")
    plt.plot([x_in, x_in], [0, y_in], c="r")
    plt.plot([-5, x_in], [y_in, y_in], c="r")
```

```python
ipywidgets.interact(plot_sigmoid, x_in=(-5, 5, 0.1))
```

```python
# plot hist gram 
def plot_hist(mu=0, sigma=1, n=100, bins=10, color="blue"):
    plt.xlim(-20, 20)
    X = np.random.normal(mu, sigma, n)
    plt.hist(X, bins=bins, color=color)
```

```python
ipywidgets.interact(plot_hist, mu=(-10, 10, 0.5), sigma=(0,10,0.1), n=(10, 1000, 1), bins=(1,100,1), color=["red", "green", "blue"])
```

```python
# interact_manual to run manually
ipywidgets.interact_manual(plot_hist, 
                    mu=(-10, 10, 0.5), 
                    sigma=(0,10,0.1), 
                    n=(10, 1000, 1), 
                    bins=(1,100,1), 
                    color=["red", "green", "blue"])
```
