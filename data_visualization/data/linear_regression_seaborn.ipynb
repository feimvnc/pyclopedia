Regression analysis is a set of statistical methods  
used for analyzing the relationships between a dependent  
variable and one or more independent variable(s).  

-Linear Regression Analysis  
-Multi-linear Regression Analysis  
-Non-linear Regression Analysis  

The relationship between variable is shown using equation:  
Y = a + bx + e  
Where:  
Y = Dependent variable  
x = independent variable  
a = intercept  
b = slope  
e = error(residual)  


```python
# start with simple data frame 
import pandas as pd 
import matplotlib.pylab as plt
import seaborn as sns 

df = pd.DataFrame({'x': [1,2,3,4,5,6,7],
                   'y': [2,4,6,8,9,10,12]})
sns.scatterplot(df, x='x', y='y')
plt.show()
```

```python
# regression line
sns.regplot(df, x='x', y='y')
plt.show()
```

```python
# use lmplot to draw regression line 
sns.lmplot(df, x='x', y='y')
plt.show()
```

```python
df = sns.load_dataset('flights')
```

```python
df.shape
```

```python
df.head()
```

```python
sns.lmplot(df, x='year', y='passengers')
```

```python
sns.set_theme()
sns.regplot(df, x='year', y='passengers')
```

```python
sns.set_theme()
# main plot is scatter plot
sns.jointplot(df, x='year', y='passengers')
```

```python
sns.set_theme()
# main plot is scatter plot, including plot line
sns.jointplot(df, x='year', y='passengers', kind='reg')
```

```python
sns.set_theme()
# main plot is scatter plot, including plot line
sns.regplot(df, x='year', y='passengers', 
            line_kws={'color': 'red'})
```

```python
sns.set_theme()
# main plot is scatter plot, including plot line
sns.regplot(df, x='year', y='passengers', 
            line_kws={'color': 'red'},
            scatter_kws={'color': 'blue', 'alpha':0.5},
            ci=99,
            marker='^'
           )
```

```python
sns.set_theme()
ax = sns.regplot(df, x='year', y='passengers', 
            line_kws={'color': 'red'},
            scatter_kws={'color': 'blue', 'alpha':0.5},
            ci=99,
            marker='^'
           )
ax.set_title("passengers vs year")
ax.set(xlabel='years', ylabel='passengers')
```

```python
# load penguins dataset
import matplotlib.pylab as plt 
import seaborn as sns 
df = sns.load_dataset('penguins')
print(df.head(), df.shape)
```

```python
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm')
```

```python
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm', 
           line_kws={'color': 'red'},
           scatter_kws={'color': 'teal'}
          )
```

```python
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm', 
           hue='species')
```

```python
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm', 
           col='species')
```

```python
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm', 
           col='species', hue='sex')
```

```python
# separate into rows
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm', 
           col='species', row='sex')
```

```python
# use marker for 3 species
sns.lmplot(df, x='bill_length_mm', y='flipper_length_mm', 
           hue='species', markers=['o', '*', '^'])
```

```python
# joint reg line
sns.jointplot(df, x='bill_length_mm', y='flipper_length_mm', 
           kind='reg')
```

```python
# pair plot 
sns.pairplot(df, kind='reg')
```

```python
# pair plot 
sns.pairplot(df, kind='reg', hue='species')
```

```python
# pair plot 
sns.lmplot(df, x='bill_length_mm', y='body_mass_g')
```

```python
# lmplot 
# order=1 : straight line
# order=2 : curve line
sns.lmplot(df, x='bill_length_mm', y='body_mass_g', order=2)
```
