```python
df1 = pd.read_csv('car-sales.csv')
df1.head
```

```python
x_label = 'Price'
y_label = 'Odometer (KM)'
y2_label = 'Doors'

factor = 0.001   # scale factor

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.scatter(df1[x_label]*factor,df1[y_label]*factor)
ax2.scatter(df1[x_label]*factor,df1[y2_label])

ax1.set(xlabel = x_label, ylabel = y_label)
ax2.set(xlabel = x_label, ylabel = y2_label)

ax1.xaxis.set_ticks_position('both')     # show rulers on all sides
ax1.yaxis.set_ticks_position('both')
ax2.xaxis.set_ticks_position('both')
ax2.yaxis.set_ticks_position('both')

ax1.minorticks_on()   # show minor rulers
ax2.minorticks_on() 

ax1.tick_params(which = 'major', length = 10, width = 1, direction = 'in')
ax2.tick_params(which = 'major', length = 10, width = 1, direction = 'in')
```

```python
# create two rows x two columns
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharey=True)
fig.subplots_adjust(hspace=0, wspace=0)

ax1.scatter(df1[x_label]*factor,df1[y_label]*factor)
ax2.scatter(df1[x_label]*factor,df1[y2_label])
ax3.scatter(df1[x_label]*factor,df1[y_label]*factor)
ax4.scatter(df1[x_label]*factor,df1[y2_label])
```
