```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# define dataset
x = np.arange(0, 25, 0.2)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize = (10, 7),
                facecolor='blue')
ax = plt.axes(projection = '3d')
ax.set_facecolor('white')

scatter_plot = ax.scatter3D(x, y, z, color='red',
                           depthshade=False,
                           cmap='autumn')

plt.colorbar(scatter_plot)
# ax.view_init(0, 0)   # 0 degree
ax.view_init(45, 0)   # 45 degree

ax.plot3D(x, y, z)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
```

### Surface plot

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.linspace(-5, 3, 70)
y = np.linspace(-5, 3, 70)

[X, Y] = np.meshgrid(x, y)

fig, ax = plt.subplots(1, 1)

Z = X**2 + Y**2
ax.contourf(X, Y, Z)
ax.set_title("Filled contour plot")
ax.set_xlabel("contour x")
ax.set_ylabel("contour y")
plt.show()

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma', alpha=0.2, antialiased=False)
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_zlabel("z label")

```
